```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.io
plt.rcParams['figure.figsize'] = [16, 8] #matlab operator set size of image
```

```python
A = imread(os.path.join('dog.jpg')) #load the image of the dog
X = np.mean(A, -1); 


U, S, VT = np.linalg.svd(X,full_matrices=True)
S = np.diag(S)

r = 500 #any number less than 1500 (Chosen rank < m)

I = np.linalg.inv(U[:r,:r]) @ U[:r,:r]
J = U[:r,:r] @ np.linalg.inv(U[:r,:r])
K = np.linalg.norm(I-np.eye(r))

print('the SVD is U=', U)
print('S=', S)
print('VT=', VT)
```

```python
print('We choose r=',r)
```

```python
#Showing U*U = I
print('So we have the identity matrix from U*U=',I)
```

```python
#Showing U*U != UU*
print('However, UU* doesnt generate an identity matrix as shown:',J)
print('with error', K)
```

```python
#error plot
C = [np.linalg.norm(U[:r,:r] @ np.linalg.inv(U[:r,:r]) - np.eye(r)) for r in range(U.shape[0])]

plt.yscale('log')
plt.plot(C);
```

```python
#compute the economy SVD
U_e, S_e, VT_e = np.linalg.svd(X,full_matrices=False)
S_e = np.diag(S_e)

print('the economy SVD is U=', U_e)
print('S=', S_e)
print('VT=', VT_e)
```

```python
#actual data value
x_0 = np.linalg.norm(A)

#relative error
RE = [abs((np.linalg.norm(U_e[:,:r] @ S_e[0:r,:r] @ VT_e[:r,:])-x_0)/x_0) for r in range(1,len(S_e))]

#plot part
plt.xlabel('r')
plt.plot(1 - np.array(RE), color='blue', label='Relative Reconstruction Error')
plt.plot(1 - (np.array(RE)**2), color='red', label='Variance')
plt.plot(np.cumsum(np.diag(S_e))/np.sum(np.diag(S_e)), color='black', label='Cumulative Sum')
plt.legend();
```

```python
print('RE recaptures the 99% of variance when r is around 20, 99% of cumulative sum when r is at most 20, although note that they have different starting points')
```

```python
mat_contents = scipy.io.loadmat(os.path.join('allFaces.mat')) #Load the Yale B image
faces = mat_contents['faces']
m = int(mat_contents['m'])
n = int(mat_contents['n'])
nfaces = np.ndarray.flatten(mat_contents['nfaces'])

allPersons = np.zeros((n*6,m*6))
count = 0

for j in range(6):
    for k in range(6):
        allPersons[j*n : (j+1)*n, k*m : (k+1)*m] = np.reshape(faces[:,np.sum(nfaces[:count])],(m,n)).T
        count += 1
        
img = plt.imshow(allPersons)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U_f, S_f, VT_f = np.linalg.svd(allPersons, full_matrices=False) #compute the economy SVD
```

```python
#Using method of snapshots

VS2VT = faces[:,:36].T @ faces[:,:36]  #XTX
```

```python
#computing SVD using MOS

S2_MOS, V_MOS = np.linalg.eig(VS2VT) #sigma squared and V from MOS
U_MOS = faces[:,:36] @ V_MOS @ np.linalg.inv(np.diag(np.sqrt(abs(S2_MOS)))) #U from MOS = X@V@Sigma from MOS inverse
```

```python
#Singular Values Comparison

plt.figure(figsize=(10,5))
plt.plot(S_f, label='SVD singular values')
plt.plot(np.sqrt(abs(S2_MOS)),  label='MoS singular values')
plt.yscale('log')
plt.legend();
```
