```python
#fourier transform
dx = 0.02
L = 2
x = L * np.arange(-1+dx,1+dx,dx)
n = 100
nquart = int(np.floor(n/4))

f = np.zeros(n)
for l in range(nquart,3*nquart - 1):
    f[l] = 1-np.abs(x[l])


G = np.concatenate([x, f])

FT = np.fft.fft(G).real

plt.plot(FT)
```
