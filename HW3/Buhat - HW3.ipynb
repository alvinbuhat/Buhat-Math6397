```python
A = imread(os.path.join('dog.jpg')) #load the image of the dog
X = np.mean(A, -1); 
plt.imshow(X, cmap='gray')
plt.axis('off');
```

```python
#Use FFT to compress the image
X_fft = np.fft.fft2(X)
```

```python
X_fs = np.sort(np.abs(X_fft.reshape(-1))) #Sort the FFTd data by its absolute value

#Computing the error
error = []

for k in range(1,101):                                
    thresh = X_fs[int(np.floor((1-k/100)*len(X_fs)))] # calculate percent threshold by iteratively increasing compression ratio(%)
    i = np.abs(X_fft)>thresh                          # set small indices i
    X_f_small = X_fft * i                             # Threshold small indices 
    X_small = np.fft.ifft2(X_f_small).real            # Recover X pic
    error.append(((X - X_small)**2).mean())           # mean squared error between compressed image and real one for every iteration
```

```python
#Plotting the error between the actual and compressed images

plt.figure(figsize=(20,5))
plt.plot(error)
plt.xlabel('compression ratio')
plt.ylabel('MSE');
```
